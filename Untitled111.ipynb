```python
# ============================================================
# 🔥 CrimeScope — High Crime Rate Area Detection
# ============================================================

import os
import json
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================================
# 📂 Paths
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\AI Crime Pattern and Hotspot Predictor"
DATA_PATH = os.path.join(BASE_DIR, "archive", "crime_dataset_india.csv")
PKL_PATH = os.path.join(BASE_DIR, "crime_model.pkl")
VIS_DIR = os.path.join(BASE_DIR, "visuals")
os.makedirs(VIS_DIR, exist_ok=True)

# ============================================================
# 📥 Load Dataset & Model
# ============================================================

print("[INFO] Loading dataset and model...")
df = pd.read_csv(DATA_PATH)
artifacts = joblib.load(PKL_PATH)

model = artifacts["model"]
scaler = artifacts["scaler"]
label_encoders = artifacts["label_encoders"]
feature_cols = artifacts["feature_cols"]
le_crime = label_encoders["crime_domain"]

print(f"[INFO] Expected Features: {feature_cols}")

# ============================================================
# 🧹 Data Cleaning & Feature Prep
# ============================================================

df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]

# ---- Parse datetime ----
if "date_of_occurrence" in df.columns:
    df["date_of_occurrence"] = pd.to_datetime(df["date_of_occurrence"], errors="coerce", dayfirst=True)
else:
    df["date_of_occurrence"] = pd.NaT

if "time_of_occurrence" in df.columns:
    df["time_of_occurrence"] = pd.to_datetime(df["time_of_occurrence"], errors="coerce", dayfirst=True)
else:
    df["time_of_occurrence"] = pd.NaT

# ---- Temporal features ----
df["year"] = df["date_of_occurrence"].dt.year.fillna(2020).astype(int)
df["month"] = df["date_of_occurrence"].dt.month.fillna(1).astype(int)
df["hour"] = pd.to_datetime(df["time_of_occurrence"], errors="coerce").dt.hour
mask = df["hour"].isna()
df.loc[mask, "hour"] = np.random.randint(0, 24, size=mask.sum())
df["hour"] = df["hour"].astype(int)

# ---- Victim age ----
if "victim_age" in df.columns:
    df["victim_age"] = pd.to_numeric(df["victim_age"], errors="coerce")
    med = df["victim_age"].median(skipna=True)
    if pd.isna(med):
        med = 30
    df["victim_age"] = df["victim_age"].fillna(med).astype(int)
else:
    df["victim_age"] = np.random.randint(18, 60, size=len(df))

# ---- Other categorical ----
for col in ["city", "victim_gender", "weapon_used", "crime_description"]:
    if col not in df.columns:
        df[col] = "Unknown"
    df[col] = df[col].astype(str).fillna("Unknown")

df["desc_len"] = df["crime_description"].apply(lambda x: len(str(x)))

# ---- Encode categorical ----
for col in ["city", "victim_gender", "weapon_used"]:
    if col in label_encoders:
        enc = label_encoders[col]
        known = set(enc.classes_)
        df[col] = df[col].apply(lambda x: x if x in known else list(known)[0])
        df[col] = enc.transform(df[col])

# ============================================================
# 🧠 Prediction
# ============================================================

for feat in feature_cols:
    if feat not in df.columns:
        df[feat] = 0

X = df[feature_cols]
X_scaled = scaler.transform(X)
y_pred = model.predict(X_scaled)
df["predicted_domain"] = le_crime.inverse_transform(y_pred)

print(f"[INFO] Predictions generated for {len(df)} records.")

# ============================================================
# 🔥 High Crime Area Analysis
# ============================================================

crime_by_city = df.groupby("city")["predicted_domain"].count().sort_values(ascending=False)

high_crime_areas = list(crime_by_city.head(10).index)
print("\n🔴 Top 10 High Crime Areas:")
print(crime_by_city.head(10))

# ============================================================
# 📊 Visualization
# ============================================================

plt.figure(figsize=(9,6))
sns.barplot(x=crime_by_city.head(10).values, y=crime_by_city.head(10).index, palette="Reds_r")
plt.title("Top 10 High Crime Rate Areas (Predicted)")
plt.xlabel("Predicted Crime Count")
plt.ylabel("City")
plt.tight_layout()
plt.savefig(os.path.join(VIS_DIR, "high_crime_areas.png"))
plt.show()

# ============================================================
# 🌍 Optional: Geo Heatmap (if lat/long columns exist)
# ============================================================

if "latitude" in df.columns and "longitude" in df.columns:
    try:
        import folium
        from folium.plugins import HeatMap

        print("[INFO] Generating geographic heatmap...")
        map_center = [df["latitude"].mean(), df["longitude"].mean()]
        crime_map = folium.Map(location=map_center, zoom_start=5)
        heat_data = df[["latitude", "longitude"]].dropna().values.tolist()
        HeatMap(heat_data, radius=8).add_to(crime_map)
        map_path = os.path.join(VIS_DIR, "crime_heatmap.html")
        crime_map.save(map_path)
        print(f"[SAVED] Heatmap: {map_path}")
    except Exception as e:
        print(f"[WARN] Could not generate map: {e}")

# ============================================================
# 💾 Save Report
# ============================================================

report = {
    "total_records_analyzed": int(len(df)),
    "total_unique_cities": int(df["city"].nunique()),
    "high_crime_areas": high_crime_areas,
    "crime_counts_top10": crime_by_city.head(10).to_dict(),
    "visuals": ["high_crime_areas.png", "crime_heatmap.html" if "latitude" in df.columns else None]
}

json_path = os.path.join(BASE_DIR, "crimescope_high_crime_areas.json")
with open(json_path, "w") as f:
    json.dump(report, f, indent=4)

print("\n✅ High Crime Area Detection Complete!")
print(f"High Crime Areas: {high_crime_areas}")
print(f"Results saved in: {VIS_DIR}")
print(f"JSON Summary: {json_path}")
```
