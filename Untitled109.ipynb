```python
# ============================================================
#  CrimeScope  Prediction + Result Visualization (Final Fixed)
# ============================================================

import os
import json
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# ============================================================
#  Paths
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\AI Crime Pattern and Hotspot Predictor"
DATA_PATH = os.path.join(BASE_DIR, "archive", "crime_dataset_india.csv")
PKL_PATH = os.path.join(BASE_DIR, "crime_model.pkl")
VIS_DIR = os.path.join(BASE_DIR, "visuals")
os.makedirs(VIS_DIR, exist_ok=True)

# ============================================================
#  Load Dataset & Model
# ============================================================

print("[INFO] Loading dataset and model...")
df = pd.read_csv(DATA_PATH)
artifacts = joblib.load(PKL_PATH)

model = artifacts["model"]
scaler = artifacts["scaler"]
label_encoders = artifacts["label_encoders"]
feature_cols = artifacts["feature_cols"]
le_crime = label_encoders["crime_domain"]

print(f"[INFO] Expected Features: {feature_cols}")

# ============================================================
#  Data Cleaning & Feature Reconstruction
# ============================================================

df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]

# ---- Parse datetime safely ----
if "date_of_occurrence" in df.columns:
    df["date_of_occurrence"] = pd.to_datetime(df["date_of_occurrence"], errors="coerce", dayfirst=True)
else:
    df["date_of_occurrence"] = pd.NaT

if "time_of_occurrence" in df.columns:
    try:
        df["time_of_occurrence"] = pd.to_datetime(df["time_of_occurrence"], errors="coerce", dayfirst=True)
    except Exception:
        df["time_of_occurrence"] = pd.to_datetime(df["time_of_occurrence"], errors="coerce", format="%H:%M")
else:
    df["time_of_occurrence"] = pd.NaT

# ---- Temporal Features (safe) ----
df["year"] = df["date_of_occurrence"].dt.year.fillna(2020).astype(int)
df["month"] = df["date_of_occurrence"].dt.month.fillna(1).astype(int)

# Fix: safely handle missing hour values (no fillna with ndarray)
df["hour"] = pd.to_datetime(df["time_of_occurrence"], errors="coerce").dt.hour
mask = df["hour"].isna()
df.loc[mask, "hour"] = np.random.randint(0, 24, size=mask.sum())
df["hour"] = df["hour"].astype(int)

# ---- Victim age ----
if "victim_age" in df.columns:
    df["victim_age"] = pd.to_numeric(df["victim_age"], errors="coerce")
    median_age = df["victim_age"].median(skipna=True)
    if pd.isna(median_age):
        median_age = 30
    df["victim_age"] = df["victim_age"].fillna(median_age).astype(int)
else:
    df["victim_age"] = np.random.randint(18, 60, size=len(df))

# ---- Fill categorical ----
for col in ["city", "victim_gender", "weapon_used", "crime_description"]:
    if col not in df.columns:
        df[col] = "Unknown"
    df[col] = df[col].astype(str).fillna("Unknown")

df["desc_len"] = df["crime_description"].apply(lambda x: len(str(x)))

# ---- Encode categorical same as training ----
for col in ["city", "victim_gender", "weapon_used"]:
    if col in label_encoders:
        enc = label_encoders[col]
        known = set(enc.classes_)
        df[col] = df[col].apply(lambda x: x if x in known else list(known)[0])
        df[col] = enc.transform(df[col])

# ============================================================
#  Prediction
# ============================================================

# Ensure all required columns exist
for feat in feature_cols:
    if feat not in df.columns:
        df[feat] = 0

X = df[feature_cols]
X_scaled = scaler.transform(X)
y_pred = model.predict(X_scaled)
df["predicted_domain"] = le_crime.inverse_transform(y_pred)

print(f"[INFO] Predictions generated for {len(df)} records.")

# ============================================================
#  Results & Visualization
# ============================================================

# --- Show prediction distribution ---
print("\n[RESULT] Predicted Crime Domain Distribution:")
print(df["predicted_domain"].value_counts())

plt.figure(figsize=(8,5))
sns.countplot(y=df["predicted_domain"], order=df["predicted_domain"].value_counts().index, palette="coolwarm")
plt.title("Predicted Crime Domain Distribution")
plt.xlabel("Count")
plt.ylabel("Crime Domain")
plt.tight_layout()
plt.savefig(os.path.join(VIS_DIR, "prediction_distribution.png"))
plt.show()

# --- If ground truth labels exist, compute accuracy ---
if "crime_domain" in df.columns:
    try:
        y_true = label_encoders["crime_domain"].transform(df["crime_domain"])
        y_pred_enc = label_encoders["crime_domain"].transform(df["predicted_domain"])
        acc = accuracy_score(y_true, y_pred_enc)
        print(f"[RESULT] Model Accuracy: {acc:.4f}")
        print(classification_report(y_true, y_pred_enc, target_names=le_crime.classes_))

        cm = confusion_matrix(y_true, y_pred_enc)
        plt.figure(figsize=(8,6))
        sns.heatmap(cm, cmap="coolwarm", annot=False)
        plt.title("Confusion Matrix Heatmap")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.tight_layout()
        plt.savefig(os.path.join(VIS_DIR, "confusion_heatmap.png"))
        plt.show()
    except Exception as e:
        print(f"[WARN] Could not compute accuracy: {e}")

# ============================================================
#  Save JSON Summary
# ============================================================

result_json = {
    "total_records": int(len(df)),
    "predicted_classes": list(le_crime.classes_),
    "class_distribution": df["predicted_domain"].value_counts().to_dict(),
    "visuals": ["prediction_distribution.png", "confusion_heatmap.png" if "crime_domain" in df.columns else None]
}

json_path = os.path.join(BASE_DIR, "crimescope_prediction_result.json")
with open(json_path, "w") as f:
    json.dump(result_json, f, indent=4)

print("\n Prediction Completed Successfully!")
print(f"Results saved in: {VIS_DIR}")
print(f"JSON Summary: {json_path}")
```
