```python
# ============================================================
# 🕵️‍♂️ CrimeScope — AI Crime Pattern & Hotspot Predictor (India Edition)
# ============================================================

import os
import json
import yaml
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet  # pip install prophet
from sklearn.preprocessing import LabelEncoder

# ============================================================
# 📂 Paths
# ============================================================

BASE_DIR = r"C:\Users\NXTWAVE\Downloads\AI Crime Pattern and Hotspot Predictor"
DATA_PATH = os.path.join(BASE_DIR, "archive", "crime_dataset_india.csv")
PKL_PATH = os.path.join(BASE_DIR, "crime_model.pkl")
VIS_DIR = os.path.join(BASE_DIR, "visuals")
os.makedirs(VIS_DIR, exist_ok=True)

# ============================================================
# 📥 Load Data & Model
# ============================================================

print("[INFO] Loading data and model...")
df = pd.read_csv(DATA_PATH)
artifacts = joblib.load(PKL_PATH)

model = artifacts["model"]
scaler = artifacts["scaler"]
label_encoders = artifacts["label_encoders"]
le_crime = label_encoders["crime_domain"]
feature_cols = artifacts["feature_cols"]

print(f"[INFO] Expected Features: {feature_cols}")

# ============================================================
# 🧹 Clean & Prepare Dataset
# ============================================================

df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]

# ---- Parse datetime columns safely ----
if "date_of_occurrence" in df.columns:
    df["date_of_occurrence"] = pd.to_datetime(df["date_of_occurrence"], errors="coerce", dayfirst=True)
else:
    df["date_of_occurrence"] = pd.NaT

if "time_of_occurrence" in df.columns:
    try:
        df["time_of_occurrence"] = pd.to_datetime(df["time_of_occurrence"], errors="coerce", dayfirst=True)
    except Exception:
        df["time_of_occurrence"] = pd.to_datetime(df["time_of_occurrence"], errors="coerce", format="%H:%M")
else:
    df["time_of_occurrence"] = pd.NaT

# ---- Temporal Features (safe) ----
df["year"] = df["date_of_occurrence"].dt.year.fillna(2020).astype(int)
df["month"] = df["date_of_occurrence"].dt.month.fillna(1).astype(int)

if pd.api.types.is_datetime64_any_dtype(df["time_of_occurrence"]):
    df["hour"] = df["time_of_occurrence"].dt.hour
else:
    try:
        df["hour"] = pd.to_datetime(df["time_of_occurrence"], errors="coerce").dt.hour
    except Exception:
        df["hour"] = np.nan

mask = df["hour"].isna()
df.loc[mask, "hour"] = np.random.randint(0, 24, size=mask.sum())
df["hour"] = df["hour"].astype(int)

# ============================================================
# 🧩 Other Feature Cleanup
# ============================================================

for col in ["victim_gender", "weapon_used", "city", "crime_description", "victim_age"]:
    if col not in df.columns:
        df[col] = "Unknown"
    df[col] = df[col].fillna("Unknown").astype(str)

# ---- Fix Victim Age ----
if "victim_age" in df.columns:
    df["victim_age"] = pd.to_numeric(df["victim_age"], errors="coerce")
    median_age = df["victim_age"].median(skipna=True)
    if pd.isna(median_age):
        median_age = 30
    df["victim_age"] = df["victim_age"].fillna(median_age).astype(int)
else:
    df["victim_age"] = np.random.randint(18, 60, size=len(df))

# ---- Derived feature ----
df["desc_len"] = df["crime_description"].apply(lambda x: len(str(x)))

# ============================================================
# 🔠 Label Encoding with Trained Encoders
# ============================================================

for col in ["city", "victim_gender", "weapon_used"]:
    if col in label_encoders:
        encoder = label_encoders[col]
        known_classes = set(encoder.classes_)
        df[col] = df[col].apply(lambda x: x if x in known_classes else list(known_classes)[0])
        df[col] = encoder.transform(df[col])

# ============================================================
# 🧠 Predict Crimes
# ============================================================

# Ensure all required columns exist
for feat in feature_cols:
    if feat not in df.columns:
        df[feat] = 0

X = df[feature_cols]
X_scaled = scaler.transform(X)
y_pred = model.predict(X_scaled)
df["predicted_domain"] = le_crime.inverse_transform(y_pred)

print(f"[INFO] Predictions generated for {len(df)} records.")

# ============================================================
# 📊 1️⃣ Result Graphs
# ============================================================

plt.figure(figsize=(8,5))
sns.countplot(y=df["predicted_domain"], order=df["predicted_domain"].value_counts().index, palette="coolwarm")
plt.title("Predicted Crime Domain Distribution")
plt.xlabel("Count")
plt.ylabel("Crime Domain")
plt.tight_layout()
plt.savefig(os.path.join(VIS_DIR, "result_crime_distribution.png"))
plt.show()

importances = model.feature_importances_
plt.figure(figsize=(8,5))
sns.barplot(x=importances, y=feature_cols, palette="magma")
plt.title("Feature Importance — Crime Domain Classifier")
plt.tight_layout()
plt.savefig(os.path.join(VIS_DIR, "feature_importance.png"))
plt.show()

# ============================================================
# 📈 2️⃣ Prediction Graph — City Crime Forecast (Prophet)
# ============================================================

city_trend = (
    df.groupby(["date_of_occurrence", "city"])
    .size()
    .reset_index(name="crime_count")
    .dropna()
)

forecast_results = {}
for city_name, sub in city_trend.groupby("city"):
    if len(sub) < 20:
        continue
    try:
        temp_df = sub.rename(columns={"date_of_occurrence": "ds", "crime_count": "y"})
        model_prophet = Prophet(yearly_seasonality=True, weekly_seasonality=True)
        model_prophet.fit(temp_df)
        future = model_prophet.make_future_dataframe(periods=7)
        forecast = model_prophet.predict(future)
        forecast_results[city_name] = forecast.tail(7)[["ds", "yhat"]].to_dict(orient="records")

        if city_name in df["city"].value_counts().head(5).index:
            plt.figure(figsize=(10,4))
            plt.plot(temp_df["ds"], temp_df["y"], label="Actual")
            plt.plot(forecast["ds"], forecast["yhat"], label="Forecast", linestyle="dashed")
            plt.title(f"Crime Forecast — {city_name}")
            plt.xlabel("Date")
            plt.ylabel("Crime Count")
            plt.legend()
            plt.tight_layout()
            plt.savefig(os.path.join(VIS_DIR, f"forecast_{city_name}.png"))
            plt.show()
    except Exception as e:
        print(f"[WARN] Forecast skipped for {city_name}: {e}")

# ============================================================
# 🕊️ 3️⃣ Safe Place Analysis
# ============================================================

city_risk = df.groupby("city")["predicted_domain"].count().sort_values()
safe_places = list(city_risk.head(10).index)

plt.figure(figsize=(8,5))
sns.barplot(x=city_risk.head(10).values, y=city_risk.head(10).index, palette="Greens_r")
plt.title("Top 10 Safest Cities (Least Predicted Crimes)")
plt.xlabel("Predicted Crime Count")
plt.ylabel("City")
plt.tight_layout()
plt.savefig(os.path.join(VIS_DIR, "safe_places.png"))
plt.show()

# ============================================================
# 💾 4️⃣ Save Updated Report
# ============================================================

report = {
    "records_analyzed": int(len(df)),
    "predicted_classes": list(le_crime.classes_),
    "top_safe_cities": safe_places,
    "forecast_results_sample": {c: v for c, v in list(forecast_results.items())[:5]}
}

JSON_PATH = os.path.join(BASE_DIR, "crimescope_predictions.json")
with open(JSON_PATH, "w") as f:
    json.dump(report, f, indent=4)

print("\n✅ CrimeScope Analysis Complete!")
print(f"Safe Cities → {safe_places}")
print(f"All results saved in: {VIS_DIR}")
print(f"JSON Summary → {JSON_PATH}")
```
